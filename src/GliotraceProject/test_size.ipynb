```python
import numpy as np
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 3, figsize=(10, 6), constrained_layout=True)

# 2D grayscale (H, W)
img_hw = np.zeros((80, 120), dtype=np.uint8)
axs[0, 0].imshow(img_hw, origin="upper", interpolation="nearest", cmap="gray")
axs[0, 0].set_title("(H,W) grayscale")

# RGB (H, W, 3)
img_rgb = np.zeros((80, 120, 3), dtype=np.uint8)
img_rgb[20:60, 30:90, 0] = 255  # red rectangle
axs[0, 1].imshow(img_rgb, origin="upper", interpolation="nearest")
axs[0, 1].set_title("(H,W,3) RGB")

# RGBA (H, W, 4)
img_rgba = np.zeros((80, 120, 4), dtype=np.uint8)
img_rgba[20:60, 30:90, 1] = 255  # green rectangle
img_rgba[20:60, 30:90, 3] = 255  # alpha
axs[0, 2].imshow(img_rgba, origin="upper", interpolation="nearest")
axs[0, 2].set_title("(H,W,4) RGBA")

# 2-channel (H, W, 2) -> should error
img_hw2 = np.zeros((80, 120, 2), dtype=np.uint8)
axs[1, 0].set_title("(H,W,2) 2-channel\n(expected error)")
try:
    axs[1, 0].imshow(img_hw2, origin="upper", interpolation="nearest")
except Exception as e:
    axs[1, 0].text(0.02, 0.5, repr(e), transform=axs[1, 0].transAxes, va="center", wrap=True)
    axs[1, 0].set_xticks([]); axs[1, 0].set_yticks([])

# Stack (T, H, W) -> pick a slice to display
img_thw = np.zeros((5, 80, 120), dtype=np.uint8)
img_thw[2, 20:60, 30:90] = 255
axs[1, 1].imshow(img_thw[2], origin="upper", interpolation="nearest", cmap="gray")
axs[1, 1].set_title("(T,H,W) slice t=2")

# Invalid (H, W, 5) -> should error
img_hw5 = np.zeros((80, 120, 5), dtype=np.uint8)
axs[1, 2].set_title("(H,W,5)\n(expected error)")
try:
    axs[1, 2].imshow(img_hw5, origin="upper", interpolation="nearest")
except Exception as e:
    axs[1, 2].text(0.02, 0.5, repr(e), transform=axs[1, 2].transAxes, va="center", wrap=True)
    axs[1, 2].set_xticks([]); axs[1, 2].set_yticks([])

for ax in axs.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```
