```python
# Once the package is installed, we can access the GlioTrace by running
from gliotrace.gliotrace_class import GlioTrace
```

```python
# Some of the ways to run GlioTrace:
import pandas as pd

# ----------------------------------------------------------------------
# To run all patients all sets, no treatment, all perturbations
# ----------------------------------------------------------------------
stackfile = [
    "stacks/exp_1_roi_1_stack.npz",
    "stacks/exp_2_roi_1_stack.npz"
    ]
metadata = pd.DataFrame(
    {
        "experiment_id": [1, 2],
        "delta_t":[0.9, 0.8]
    }
)

fcols = ["vascular_distance"] # <---- for the logistic model in the hmm, 
                              #       can be changed after tracking, see 
                              #       below in the gliobj.fit_hmm() section
            

gliobj = GlioTrace(stackfile=stackfile, 
                             metadata=metadata,
                             fcols=fcols)

# ----------------------------------------------------------------------
# To run all patients all sets, no treatment, some perturbation 
# ----------------------------------------------------------------------
stackfile = [
    "stacks/exp_1_roi_1_stack.npz",
    "stacks/exp_2_roi_1_stack.npz"
    ]
metadata = pd.DataFrame(
    {
        "experiment_id": [1, 2],
        "delta_t": [0.9, 0.8],
        "perturbation": ["ctrl", "radiation"]
    }
)

fcols = ["vascular_distance"]
control = "ctrl" # <---- Filter on control perturbation

gliobj = GlioTrace(stackfile=stackfile, 
                             metadata=metadata,
                             fcols=fcols,
                             control=control # <--- Added here
                             )

# ----------------------------------------------------------------------
# To run some patients, all sets, no treatment, all perturbations
# ----------------------------------------------------------------------
stackfile = [
    "stacks/exp_1_roi_1_stack.npz",
    "stacks/exp_2_roi_1_stack.npz",
    "stacks/exp_3_roi_1_stack.npz"
    ]
metadata = pd.DataFrame(
    {
        "experiment_id": [1, 2, 3],
        "delta_t": [0.9, 0.8, 0.8],
        "patient_id": ["patient_x", "patient_y", "patient_z"]
    }
)

fcols = ["cos_sim"]
patient_id = ["patient_x", "patient_y"] # <---- Filter by patient

gliobj_3013_ctrl = GlioTrace(stackfile=stackfile, 
                             metadata=metadata, 
                             fcols=fcols,
                             patient_id=patient_id, # <---- Added here
                             )

# ----------------------------------------------------------------------
# To run all patients, subset on set, no treatment, all perturbation
# ----------------------------------------------------------------------
stackfile = [
    "stacks/exp_1_roi_1_stack.npz",
    "stacks/exp_2_roi_1_stack.npz",
    "stacks/exp_3_roi_1_stack.npz",
    "stacks/exp_4_roi_2_stack.npz"
    ]
metadata = pd.DataFrame(
    {
        "experiment_id": [1, 2, 3, 4],
        "delta_t": [0.9, 0.8, 0.8, 0.7],
        "patient_id": ["patient_x", "patient_x", "patient_x", "patient_y"],
        "set": [1, 2, 3, 1]
    }
)

fcols = ["vascular_distance"]
sets_by_patient = {
    "patient_x": [2, 3] # <---- This says that for 'patient_x' the sets 2 and 3 should be used, 
                        #       for everyone else every set
}

gliobj = GlioTrace(stackfile=stackfile, 
                             metadata=metadata, 
                             fcols=fcols,
                             sets_by_patient=sets_by_patient, # <--- added here
                            )



# ----------------------------------------------------------------------
# To run some patients, some sets, no treatment, all perturbations
# ----------------------------------------------------------------------
stackfile = [
    "stacks/exp_1_roi_1_stack.npz",
    "stacks/exp_2_roi_1_stack.npz",
    "stacks/exp_3_roi_1_stack.npz",
    "stacks/exp_4_roi_2_stack.npz"
    ]
metadata = pd.DataFrame(
    {
        "experiment_id": [1, 2, 3, 4],
        "delta_t": [0.9, 0.8, 0.8, 0.7],
        "patient_id": ["patient_x", "patient_x", "patient_y", "patient_z"],
        "set": [1, 2, 1, 2]
    }
)

fcols = ["cos_sim"]


patient_id = ["patient_x", "patient_y"]
sets_by_patient = {
    "Patient_x": [1] # <-- Limit the sets of 'patient_x' but no limit for 'patient_y'
}

# NOTE: If you want to specify BOTH patient_id and sets, then all patients in sets_by_patient must be in patient_id

gliobj = GlioTrace(stackfile=stackfile, 
                             metadata=metadata, 
                             fcols=fcols,
                             patient_id=patient_id,          # <--- Here
                             sets_by_patient=sets_by_patient # <--- Here
                            )


# ----------------------------------------------------------------------
# To run all patients, sets. Including treatment
# ----------------------------------------------------------------------
# NOTE: The below can further be combined with the filterings listed so far


stackfile = [
    "stacks/exp_1_roi_1_stack.npz",
    "stacks/exp_2_roi_1_stack.npz",
    "stacks/exp_3_roi_1_stack.npz",
    "stacks/exp_4_roi_2_stack.npz"
    ]
metadata = pd.DataFrame(
    {
        "experiment_id": [1, 2, 3, 4],
        "delta_t": [0.9, 0.8, 0.8, 0.7],
        "perturbation": ["ctrl", "radiation", "ctrl", "drug"],
        "dose": ["None", "0.5", "None", "half"]
    }
)

fcols = ["is_treatment", "vascular_distance_treat", "vascular_distance"]


# To specify treatment there are two possibilites:
treatment = ["drug", None]   # <---- This says "drug" is the treatment and no filtering on dose 
                             #       (hence 'dose' does NOT need to be a column in the metadata 
                             #        for this to be valid)
treatment = ["drug", "half"] # <---- This says "drug" is the treatment and dose = "half"

# Whenever treatment is specified, control must be specified as well:
control = "ctrl"

gliobj_3013_ctrl = GlioTrace(stackfile=stackfile, 
                             metadata=metadata, 
                             fcols=fcols,
                             treatment=treatment, # <--- Here
                             control=control,     # <--- Here
                )

# ---------------------------------------
# Question that may arise (not exhaustive):
# ----------------------------------------

# The stackfile can also be a txt file specifying the paths.
# To extract experiment id and roi number both of the patterns "exp_{integer}_" and "roi_{integer}_"
# have to be present

# The metadata can also be path to a csv file containing the same information.
# You only need one row for each experiment, otherwise the data is duplicated.


# patient_id are all assumed to be strings
# sets are ints and require a patient_id associated

# Treatment and dosage are strings.
# Treatment and perturbation are both based on the "perturbation" column in metadata, dose on 'dose' column.
# Not specifying perturbation means no filtering there.

# The program is written with lots of custom error messages which hopefully elucidates any errors met at
# the point of initalizing GlioTrace
```

```python
# Once a suitable intialization of GlioTrace has been chosen, the first step is track the cells:

# Runs tracking (note that the results are saved internally in the object)
gliobj.run_tracking()

# If save_point is specified then a GlioTrace object save_point is generated after each succesfully tracked stack
# This allows for running a subset of stacks one day and continue the next. Or if the program crashes to continue at
# the last succesfully tracked
gliobj.run_tracking(save_point="save_dir/run1")

# To resume a saved_point tracking after partial tracking use
gliobj.run_tracking(save_point="save_dir/run1", load_point="savedir/run1")

# If the signal of cells in a roi is poor, the detection algorithm can fail to find tracks. In such a case
# the stack is thrown away. If you still want to try again, use the detection_backup to retry finding cells
# with a greater sensitivity
gliobj.run_tracking(detection_backup=0.4)
```

```python
# After run_tracking we can visualize the reults for any tracked stack:
gliobj.print_exp_roi() # <---- Tells us which experiments and rois we have

exp = 1
roi = 1
gliobj.video_tracking(exp=exp,roi=roi) # <--- Visualize the tracking for a particular exp, roi, writes a mp4
```

```python
# To run the hmm model we call
gliobj.fit_hmm()

# We can change the columns that are used in the logistic regression without re-tracking the stacks
gliobj.fit_hmm(fcols=["speed"])
```

```python
# The objects which the hmm generates can be accessed as follows:
gliobj.data_feat # The data which went into the hmm
gliobj.gammas # Posterior class probabilites 
gliobj.pi # Inital distribution for each of the classes
gliobj.glm_models # The sklearn LogisticRegression object fitted on the best likelihood E/M step
gliobj.transition_matrix # The global transition matrix

# After the hmm, viterbipaths are also created. These can be visualzied using
exp=1
roi=1
gliobj.video_compare(exp=exp, roi=roi)
```

```python
# After running the hmm it is also possible to generaty summary statistics:

# Ex 1: Metadata does not have patient id, no treatment specified then
# the following generates a dictionary with keys that are the experiment_id
# and values which are time x roi summaries of the fcol using the mode
ex1 = gliobj.generate_summary_statistics(fcol="speed", mode="mean")

# Ex 2: Treatment is present
ex2 = gliobj.generate_summary_statistics(fcol="speed", treatment=False, mode="mean")
ex3 = gliobj.generate_summary_statistics(fcol="speed", treatment=True, mode="mean")
ex4 = gliobj.generate_summary_statistics(fcol="speed", mode="mean") # <--- both

# Ex 3: Metadata has patient id not sets and no treatment
ex5 = gliobj.generate_summary_statistics(fcol="speed", patient_id="patientx", mode="mean")

# Ex 4: Metadata has patient id and sets, no treat
ex5 = gliobj.generate_summary_statistics(fcol="speed", patient_id="patientx", set_id=1, mode="mean")

# Whenever patient_id and sets are in the metadata column, the ones which are used in the program can be listed as
gliobj.patients()                       # <---- List all patients
gliobj.sets(patient="patient_x")        # <---- List all sets for the given patient
gliobj.exps(patient="patient_x", set=1) # <---- List all exps for patient and set

# We can also filter by patients, sets, exp which are treated or untreated:
gliobj.patients(treated=True)      
gliobj.sets(patient="patient_x", treated=True) 
gliobj.exps(patient="patient_x", set=1, treated=True)
```

```python
# Save loading and object details

# At anytime you can list the object details by running
gliobj.print_configuration()

# To see the subtable which lists the filepaths, experiment_id, if things are tracked etc
gliobj.subtable()

# To save the object at any point we can do
gliobj.save_run("run_dir/run001")

# To initalize a new GlioTrace object from a savepoint:
new_gliobj = GlioTrace.load_run("run_dir/run001")
```
